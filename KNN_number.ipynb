```python
import numpy as np
import pandas as pd
import math

# 전체 data 읽어 오기
df = pd.read_csv('iris_KNN.csv', encoding= 'utf-8', header=None, skiprows = None, nrows= None)
```

k최근접분류기
각 데이터간 거리를 측정한다.
가장 짧은 데이터 레이블로 k_label로 새로 지정한다.

각 데이터 별로 거리값을 np.array로 저장한다. [결과1, 결과2, ...0, ]
거리값이 0이면 그게 자신의 인덱스다.
array 전체를 이차원으로 저장한다.

```python
# 100개를 임의 추출, 항상 같은 33부터 시작.
sample_df = df.sample(n=100, random_state=33)

# test를 전체 df에서 sample 지워서 만든다.
test_df = df.drop(sample_df.index)

print(sample_df.head(3))
# 거리 계산 함수
def calculate_distance(p1, p2):
    return math.sqrt(sum( (d1 - d2)**2 for d1, d2 in zip(p1, p2)))

sample_df = sample_df.reset_index(drop=False)
sample_df.columns = ['oldIndex','f1', 'f2', 'f3', 'f4','label']

print(sample_df.head(3))

# 거리 계산을 위한 df

sample = sample_df
calc_df = sample.loc[:,['f1', 'f2', 'f3', 'f4']]

# 최단 거리 행 인덱스 리스트

# 각 데이터 거리결과값 np.array 이차원으로 저장
dist_arr = np.array([], dtype=object)

# 한 점과 다른 모든 점과 거리를 계산 후 리스트로 모은다.
for i in range(len(sample)):
   # 두 점 간 거리 리스트
    p1 = calc_df.iloc[i]
    dist = np.array
    for j in range(len(sample)):
        p2 = calc_df.iloc[j]
        distance = calculate_distance(p1, p2)
        dist[j] = distance

 # 한 점씩 결과 리스트를 series로 만들어 argmin으로 최소값 인덱스를 구한다.
    np.put(dist_arr, i, dist)


smallest = []
smallest.append(dist.argmin())

# 최소거리 행 인덱스를 컬럼으로 추가한다.
smallest = pd.Series(smallest)
sample.insert(5, 'knn_1', smallest)

print(sample.head())

# label 과 knn_1 둘 다 중복값을 갖는 데이터
# 새로 할당할 레이블 리스트를 컬럼으로 만든다.
klabel = []

for i in range(len(sample)):
    kid = sample.loc[i, 'knn_1']
    lb = sample.loc[kid, 'label']
    klabel.append(lb)

klabel = pd.Series(klabel)
sample_df.insert(6, 'k1_label', klabel)
print('\nk=1 일 때, 가장 가까운 거리값을 가진 행 레이블을 k1_label 로 할당함.')
sample_df = sample.sort_values(by=['oldIndex'])
print(sample_df.head())

# 새로 지정한 k1_label 별로 count
print('\n학습 데이터 100개 중 k1_label 로 그룹화해서 갯수를 구함')
print(sample_df.groupby(by=['k1_label']).count())
# 원래 설정된 label count
print(sample_df.groupby(by=['label']).count())

```
