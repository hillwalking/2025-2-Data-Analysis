```python
import numpy as np
import pandas as pd
import math
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns


## 시각화 패키지와 한글 폰트 지정
plt.rc('font', family = 'NanumBarunGothic')

# 전체 data 읽어 오기
df = pd.read_csv('iris_KNN.csv', encoding= 'utf-8', header=None, skiprows = None, nrows= None)

# 인덱스 확인
df.index

# 컬럼명 지정 :  특징 4가지, 클래스레이블
df.columns  = ['f1', 'f2', 'f3', 'f4', 'label']

# 100개를 임의 추출, 항상 같은 33부터 시작.
sample_df = df.sample(n=100, random_state=33)

# test를 전체 df에서 sample 지워서 만든다.
test_df = df.drop(sample_df.index)

# 원래 label 열과 비교할 수 있도록, 판별 결과를 저장할 컬럼을 만들어 둔다.
sample_df.loc[: , 'class_id'] = 0
test_df.loc[:, 'class_id'] = 1

# 특징 컬럼만 이용
sample_df = sample_df.reset_index(drop='true')
f_df = sample_df.loc[:, ['f1', 'f2', 'f3', 'f4']]
sample_df.head()


sample_df.loc[10]
```

```python
# 두 데이터 간 거리 계산
def euclidean_distance( pt1, pt2 ):
    return math.sqrt( sum( [ ( d1 - d2 )**2 for d1, d2 in zip( pt1, pt2 ) ] ) )
# 각 점에서 거리를 계산한 결과 리스트에서 최소거리를 찾아 인덱스를 반환
def classify_label(dic_list):
    min_dist = 100
    min_id = 0
    for id, dist in enumerate(dic_list):
        # print(id,':', dist)
        if(min_dist > dist):
            min_dist = dist
            min_id = id
    return min_id

# print(f_df.head(10))
for i in range(len(f_df)):
    pt1 = f_df.iloc[i]
    dic_list = []
    for j in range(len(f_df)):
        if(i != j):
            pt2 = f_df.iloc[j]
            dic_list.append(euclidean_distance(pt1, pt2))
    min_id = classify_label(dic_list)
    sample_df[i]['class_id'] = sample_df.loc[min_id]['label']

sample_df.head(20)


#     if(minval > val):
#         minval = val
#         id = key
# print(id, ': ', minval)
#                 # f_df['result_label'] =  sample_df.iloc[sample_df.index==j]['clabel']
#                 print('min_dist', min_dist, '\n')



```
